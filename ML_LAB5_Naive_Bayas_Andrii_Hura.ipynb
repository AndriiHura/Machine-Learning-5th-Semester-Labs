from sklearn import datasets
from sklearn import metrics
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
import matplotlib.pyplot as plt

```python
def split_data(percent,data, target):
    """Function splits data according to percent
    percent specifies percent of test set"""
    #np.random.seed(1)
    indices = np.random.permutation(len(data))
    n_training_samples = int((len(data)/100)*percent)
    train_set_data = data[indices[:-n_training_samples]]
    train_set_labels = target[indices[:-n_training_samples]]
    test_set_data = data[indices[-n_training_samples:]]
    test_set_labels = target[indices[-n_training_samples:]]
    result = [train_set_data,train_set_labels,test_set_data,test_set_labels]
    return result


def test_accuracy(actual, predicted):
    counter = 0
    for ind in range(len(actual)):
        if(actual[ind]==predicted[ind]):
            counter +=1
    return (counter/len(actual))*100
```
